```python
import seaborn as sns
```

```python
data.hist(bins=20, figsize=(12, 12))
plt.show()
```

```python
column_of_interest = 'pH'
Q1 = data[column_of_interest].quantile(0.25)
Q3 = data[column_of_interest].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data[column_of_interest] < lower_bound) | (data [column_of_interest] > upper_bound)]

plt.figure(figsize=(12, 8))
sns.boxplot(x=data[column_of_interest])
plt.title(f'Boxplot of {column_of_interest} with Outliers Highlighted')
plt.scatter(outliers.index, outliers[column_of_interest], color='red', label='Outliers')
plt.legend()
plt.show()
```
